```
import numpy as np 
import pandas as pd 
```

```
crx_data=pd.read_csv("/content/crx.data.csv")
crx_data.head()
```

```
crx_data.shape
```

### **TASK 1:**

**In the dataset crx.data, name the columns as col1, col2..col16**

```
string="col"
col_list=[]
for i in range(1,17):
  new_string=string+str(i)
  col_list.append(new_string)

crx_data.columns=col_list
crx_data.head()
```

### **TASK 2:**

**Print the last 10 rows of data**

```
crx_data.tail(10)
```

### **TASK 3:**
**Replace the ‘?’ with Not-a-Number**

```
for cols in col_list:
  crx_data[crx_data[cols]=='?']=np.nan 
crx_data.head()
```

### **TASK 4:** 
**Comment on the datatype of variables**

```
crx_data.info()
```

**Inference :** 

There are 16 features in which **4 features** are categorical features(datatype :Object) and remaining **12 features** are numerical features (datatype : float64)

### **TASK 5:**
**The col16 has + and -, replace them ‘P’ and ‘N’ respectively**

```
crx_data['col16'].value_counts()
```

```
crx_data=crx_data.replace({'col16': {'+': 'P', '-': 'N'}})
crx_data.head()	
```
